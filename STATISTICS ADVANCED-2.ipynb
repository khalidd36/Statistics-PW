## Question 1: What is hypothesis testing in statistics?

**Answer:**
Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves testing an assumption (hypothesis) about a population parameter and determining whether sample evidence provides enough support to reject or fail to reject that assumption.

---

## Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

**Answer:**

* **Null hypothesis (H₀):** A statement of no effect, no difference, or status quo (e.g., the population mean equals 50).
* **Alternative hypothesis (H₁ or Ha):** A statement that contradicts the null hypothesis, indicating the presence of an effect or difference (e.g., the population mean is not equal to 50).

The null hypothesis is assumed true until evidence suggests otherwise; the alternative is what researchers typically aim to support.

---

## Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

**Answer:**
The **significance level (α)** is the probability threshold for rejecting the null hypothesis, commonly set at 0.05 (5%). It represents the maximum acceptable probability of making a **Type I error** (rejecting a true null). If the p-value from the test is less than α, we reject H₀; otherwise, we fail to reject H₀.

---

## Question 4: What are Type I and Type II errors? Give examples of each.

**Answer:**

* **Type I error:** Rejecting H₀ when it is actually true. Example: Concluding a drug is effective when it is not.
* **Type II error:** Failing to reject H₀ when it is false. Example: Concluding a drug has no effect when it actually does.

---

## Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

**Answer:**

* **Z-test:** Used when the population variance is known or the sample size is large (n > 30). It assumes normal distribution.
* **T-test:** Used when the population variance is unknown and the sample size is small (n ≤ 30). It uses the t-distribution with degrees of freedom.

**When to use:**

* Z-test: Large samples, known σ.
* T-test: Small samples, unknown σ.

Question 6: Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n, p = 10, 0.5
size = 1000

# Generate binomial random numbers
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(-0.5, n+1.5, 1), edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


```python

from math import sqrt

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

sample = np.array(sample_data)
n = len(sample)
mean_sample = sample.mean()
std_sample = sample.std(ddof=1)

# Hypothesis: H0: mu = 50
mu0 = 50
se = std_sample / sqrt(n)

z_stat = (mean_sample - mu0) / se
print("Sample mean:", mean_sample)
print("Z-statistic:", z_stat)

# Interpretation:
# If |z| > 1.96 (for alpha=0.05, two-tailed), reject H0.
```

Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


```python
from math import sqrt
from scipy import stats

# Generate normal data
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=100)

# Sample mean and std
data_mean = np.mean(data)
data_std = np.std(data, ddof=1)
n = len(data)

# Standard error
se = data_std / sqrt(n)

# 95% CI
t_crit = stats.t.ppf(0.975, df=n-1)
ci_lower = data_mean - t_crit * se
ci_upper = data_mean + t_crit * se

print(f"Sample mean: {data_mean:.3f}")
print(f"95% CI: [{ci_lower:.3f}, {ci_upper:.3f}]")

# Plot data
plt.hist(data, bins=15, edgecolor='black')
plt.title("Normal Distribution Simulation")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.


```python
# Example dataset
data = [10, 12, 9, 15, 14, 10, 11, 13, 12, 16]
data = np.array(data)

# Calculate Z-scores
mean_data = np.mean(data)
std_data = np.std(data, ddof=1)
z_scores = (data - mean_data) / std_data

print("Z-scores:", z_scores)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=5, edgecolor='black')
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
